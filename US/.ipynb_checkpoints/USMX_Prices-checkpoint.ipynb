# US market analysis

```python
# Importing libraries 
import pandas as pd 
import numpy as np
from datetime import date
```

```python
class crop():
    
    def __init__(self, name, name_abb, conv_package_un, cutoff):
        self.name = name
        self.name_abb = name_abb
        self.conv_package_un = conv_package_un
        self.cutoff = cutoff
        
    def get_name(self):
        return self.name
    
    def get_name_abb(self):
        return self.name_abb
    
    def get_conv_package_un(self):
        return self.conv_package_un

    def get_data(self):
        return self.data
    
    def current_day(self):
        return date.today().strftime('%d')
    
    def current_month(self):
        return date.today().strftime('%m')   
     
    def current_year(self):
        return date.today().strftime('%Y')
    
    def cutoff_day(self):
        return self.cutoff[-2:]
    
    def cutoff_month(self):
        return self.cutoff[4:6] 
     
    def cutoff_year(self):
        return self.cutoff[:4] 
    
    def lb_to_kg(self):
        return 0.453592   
    
    def read_prices(self):
        
        # Date to : current date data to collect updated information
        tday = date.today().strftime('%d')
        tmonth = date.today().strftime('%m')
        tyear = date.today().strftime('%Y')
        
        # URL for accessing prices
        USDA = f"https://www.marketnews.usda.gov/mnp/fv-report-top-filters\
?&commAbr={self.name_abb}&varName=&locAbr=&repType=shipPriceDaily\
&navType=byComm&locName=&navClass=&type=shipPrice&dr=1&volume=&commName={self.name}&navClass,=&portal=fv&region=\
&repDate={self.cutoff[4:6]}%2F{self.cutoff[-2:]}%2F{self.cutoff[:4]}\
&endDate={tmonth}%2F{tday}%2F{tyear}&format=excel&rebuild=false"
         
        # Assign the table data in html format to a Pandas dataframe
        table =  pd.read_html(USDA,header=0,parse_dates=['Date'])[0]

        # Store the dataframe in Excel file to avoid html tags and save data source file
        file_path = f'./Data/{self.name}/USMXprices.xls'
        table.to_excel(file_path) 

        # Read the Excel in a Pandas dataframe and save it in a new attribute called data
        self.data = pd.DataFrame()
        self.data = pd.read_excel(file_path\
                            ,header=0
                            ,usecols= (['Commodity Name',\
                            'City Name',\
                            'Package',\
                            'Date',\
                            'Low Price',\
                            'High Price',\
                            'Mostly Low',\
                            'Mostly High',\
                            'Season',\
                            'Type']))
        return self.data.head(1)
    
    def keep_single_city(self, city):   
        self.data.drop(self.data[self.data['City Name'].astype(str).str[:6] != city].index, inplace = True)
        return self.data.head(1)
    
    def list_cities(self):   
        return self.data.groupby('City Name')['Date'].nunique()
    
    def delete_fruit_type(self, fruit_type):   
        self.data.drop(self.data[self.data['Type'] == fruit_type].index, inplace = True)

    def list_fruit_types(self):   
        return self.data.groupby('Type')['Date'].nunique()
    
    def calc_avg_price(self):   
        self.get_data()['Avg Price'] = np.where(self.get_data()['Mostly High'].notnull(),\
                                          self.get_data()[['Mostly Low','Mostly High']].mean(axis=1),\
                                          self.get_data()[['Low Price','High Price']].mean(axis=1))
        # https://stackoverflow.com/questions/36909977/update-row-values-where-certain-condition-is-met-in-pandas/36910033
        self.get_data()['Avg Price'].dropna(inplace = True)  
       
```

```python
class campaign():
    
    def __init__(self, name, city, dates):
        self.name = name
        self.city = city
        self.dates = dates
        self.weeks = 10  #diff dates
```

```python
blue_mxn_prices = crop('BLUEBERRIES','BLUBY', 72 * 0.0283495,'20180101')
blue_mxn_prices.read_prices()
blue_mxn_prices.list_cities()
blue_mxn_prices.keep_single_city('MEXICO')
blue_mxn_prices.list_cities()
blue_mxn_prices.delete_fruit_type('Organic')
blue_mxn_prices.list_fruit_types()
blue_mxn_prices.calc_avg_price()
```
